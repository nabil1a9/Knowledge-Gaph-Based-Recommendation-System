```python
from sentence_transformers import SentenceTransformer, util
from itertools import combinations

# Load SBERT model
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')



# Function to generate n-grams
def find_ngrams(input_list, n):
    return list(zip(*[input_list[i:] for i in range(n)]))

# Calculate embeddings for n-grams
def get_synonyms(title, content):
    words_title = title.lower().split()
    print(content)
    print('\n')
    words_content = content.lower().split()

    potential_synonyms = []
    for n in range(1, min(len(words_title), len(words_content)) + 1):
        ngrams_title = find_ngrams(words_title, n)
        ngrams_content = find_ngrams(words_content, n)

        embeddings_title = model.encode([' '.join(gram) for gram in ngrams_title], convert_to_tensor=True)
        embeddings_content = model.encode([' '.join(gram) for gram in ngrams_content], convert_to_tensor=True)

        for i, ngram_embedding in enumerate(embeddings_title):
            for j, other_ngram_embedding in enumerate(embeddings_content):
                similarity = util.pytorch_cos_sim(ngram_embedding, other_ngram_embedding)
                if similarity > 0.8:  # Adjust the threshold as needed
                    potential_synonyms.append([list(ngrams_title[i]), list(ngrams_content[j])])

    # Flatten the list of synonyms and convert to tuples
    flattened_synonyms = [tuple(word) for sublist in potential_synonyms for word in sublist]

    # Remove duplicates
    unique_synonyms = list(set(flattened_synonyms))
    return unique_synonyms

# Apply the synonym generation process to the DataFrame
DF['lemma_Title'] = DF['lemma_Title'].str.replace(r'\d+\.\d+', '', regex=True)
DF['lemma_Content'] = DF['lemma_Content'].str.replace(r'\d+\.\d+', '', regex=True)

# Apply the synonym-finding code
DF['lemma_Title'] = DF['lemma_Title'].str.replace(r'\d', '', regex=True)
DF['lemma_Content'] = DF['lemma_Content'].str.replace(r'\d+(\.\d+)?', '', regex=True)


DF['lemma_Title'] = DF['lemma_Title'].fillna('')
DF['lemma_Content'] = DF['lemma_Content'].fillna('')

DF['Synonyms']= DF.apply(lambda x: get_synonyms(x['lemma_Title'], x['lemma_Content']), axis=1)
```

```python
# Generate
gen_kwargs = {
    "max_length": 216,
    "length_penalty": 0,
    "num_beams": 5,
    "num_return_sequences": 4
}

def extract_relations_from_model_output(text):
    relations = []
    relation, subject, relation, object_ = '', '', '', ''
    text = text.strip()
    current = 'x'
    text_replaced = text.replace("<s>", "").replace("<pad>", "").replace("</s>", "")
    for token in text_replaced.split():
        if token == "<triplet>":
            current = 't'
            if relation != '':
                relations.append({
                    'head': subject.strip(),
                    'type': relation.strip(),
                    'tail': object_.strip()
                })
                relation = ''
            subject = ''
        elif token == "<subj>":
            current = 's'
            if relation != '':
                relations.append({
                    'head': subject.strip(),
                    'type': relation.strip(),
                    'tail': object_.strip()
                })
            object_ = ''
        elif token == "<obj>":
            current = 'o'
            relation = ''
        else:
            if current == 't':
                subject += ' ' + token
            elif current == 's':
                object_ += ' ' + token
            elif current == 'o':
                relation += ' ' + token
    if subject != '' and relation != '' and object_ != '':
        relations.append({
            'head': subject.strip(),
            'type': relation.strip(),
            'tail': object_.strip()
        })
    return relations
class KB():
    def __init__(self):
        self.relations = []

    def are_relations_equal(self, r1, r2):
        return all(r1[attr] == r2[attr] for attr in ["head", "type", "tail"])

    def exists_relation(self, r1):
        return any(self.are_relations_equal(r1, r2) for r2 in self.relations)

    def add_relation(self, r):
        if not self.exists_relation(r):
            self.relations.append(r)

    def print(self):
        print("Relations:")
        for r in self.relations:
            print(f"  {r}")


    def create_df(self):
        reln_df = pd.DataFrame(self.relations)
        return reln_df
def from_small_text_to_kb(text, verbose=False):
    kb = KB()

    # Tokenizer text
    model_inputs = tokenizer(text,
                             max_length=512,
                             padding=True,
                             truncation=True,
                            return_tensors='pt')
    if verbose:
        print(f"Num tokens: {len(model_inputs['input_ids'][0])}")

    # Generate
    gen_kwargs = {
        "max_length": 216,
        "length_penalty": 0,
        "num_beams": 3,
        "num_return_sequences": 3
    }
    generated_tokens = model.generate(
        **model_inputs,
        **gen_kwargs,
    )
    decoded_preds = tokenizer.batch_decode(generated_tokens,
                                           skip_special_tokens=False)

    # create kb
    for sentence_pred in decoded_preds:
        relations = extract_relations_from_model_output(sentence_pred)
        for r in relations:
            kb.add_relation(r)

    return kb
```

# Semantic search between concept and triple objects to extract similar concepts.

```python
! pip install -U sentence-transformers
from sentence_transformers import SentenceTransformer, util
import torch
import pandas as pd

embedder = SentenceTransformer('all-MiniLM-L6-v2')

# Corpus with example sentences
corpus = list(DFTRIPLE2['head'])
corpus3 = list(DFTRIPLE2['index'])
corpus2 = list(DFTRIPLE2['tail'])
corpus1 = list(DFTRIPLE2['type'])

corpus_embeddings = embedder.encode(corpus, convert_to_tensor=True)

# Query sentences:
queries = list(DF['lemma_Title'])

# Find the closest 5 sentences of the corpus for each query sentence based on cosine similarity
result_df = pd.DataFrame(columns=['Title', 'Head', 'index', 'type', 'tail'])

top_k = min(200, len(corpus))
for query in queries:
    query_embedding = embedder.encode(query, convert_to_tensor=True)

    # Calculate cosine similarity scores
    cos_scores = util.cos_sim(query_embedding, corpus_embeddings)[0]

    # Filter the results with a similarity threshold of 0.6
    similar_indices = [idx for idx, score in enumerate(cos_scores) if score > 0.6]

    # Sort the indices by score in descending order
    similar_indices.sort(key=lambda x: cos_scores[x], reverse=True)

    # Take the top results or all available if there are fewer
    top_results = similar_indices[:top_k]

    for idx in top_results:
        result_df = result_df.append({
            'Title': query,
            'Head': corpus[idx],
            'index': corpus3[idx],
            'tail': corpus2[idx],
            'type': corpus1[idx]
        }, ignore_index=True)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk

# Load your DataFrame (replace this with your data)
# Assuming you have a DataFrame called 'df' with columns 'head', 'tail', 'type', and 'content'
# Example:
# df = pd.read_csv('your_data.csv')
df=merged_new_concept
# Define the TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the content to get TF-IDF values for each word
tfidf_matrix = tfidf_vectorizer.fit_transform(merged_new_concept['Content'])

# Get the TF-IDF matrix as a DataFrame with feature names
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

# Set a TF-IDF threshold (you can adjust this threshold)
threshold = 0.6

# Create a new DataFrame to store the filtered titles
filtered_df = pd.DataFrame(columns=['head', 'tail', 'type'])

# Tokenize the titles using NLTK
nltk.download('punkt')
from nltk.tokenize import word_tokenize

# Iterate through each title and check if it meets the threshold
for index, row in df.iterrows():
    title = row['head']  # Get the title from the 'head' column
    title_tokens = word_tokenize(title)  # Tokenize the title
    title_tfidf = sum(tfidf_df[token].iloc[index] for token in title_tokens if token in tfidf_df)

    if title_tfidf >= threshold:
        filtered_df = filtered_df.append({'head': row['head'], 'tail': row['tail'], 'type': row['type']}, ignore_index=True)

# Display the resulting DataFrame
filtered_df
```

## Concat similar Concept in one List

```python
import pandas as pd
from sentence_transformers import SentenceTransformer, util

# Initialize the model
model = SentenceTransformer('multi-qa-MiniLM-L6-cos-v1')
def encode_text(text):
    return model.encode(text, convert_to_tensor=True)
# Sample DataFrame

df = filtered_df


# Create an empty DataFrame to store the results
result_df = pd.DataFrame(columns=['head_list', 'type_list', 'tail_list'])

# Set a similarity threshold
similarity_threshold = 0.8
# Iterate through the rows of the original DataFrame
for index, row in df.iterrows():
    print(index)
    current_head = row['head']
    head_list = [current_head]
    tail_list = [row['tail']]
    type_list = [row['type']]

    # Iterate through the rows in the result DataFrame
    for _, result_row in result_df.iterrows():
        similarity_score = util.pytorch_cos_sim(encode_text(current_head), encode_text(result_row['head_list'][0]))

        if similarity_score > similarity_threshold:
            head_list.extend(result_row['head_list'])
            tail_list.extend(result_row['tail_list'])
            type_list.extend(result_row['type_list'])

    if head_list != [current_head]:
        # Update the result DataFrame
        result_df.loc[result_df['head_list'].apply(lambda x: x[0] in head_list), 'head_list'] = result_df.loc[result_df['head_list'].apply(lambda x: x[0] in head_list), 'head_list'].apply(lambda x: x + head_list)
        result_df.loc[result_df['head_list'].apply(lambda x: x[0] in head_list), 'tail_list'] = result_df.loc[result_df['head_list'].apply(lambda x: x[0] in head_list), 'tail_list'].apply(lambda x: x + tail_list)
        result_df.loc[result_df['head_list'].apply(lambda x: x[0] in head_list), 'type_list'] = result_df.loc[result_df['head_list'].apply(lambda x: x[0] in head_list), 'type_list'].apply(lambda x: x + type_list)
    else:
        # Create a new row
        result_df = result_df.append({'head_list': head_list, 'tail_list': tail_list, 'type_list': type_list}, ignore_index=True)
```

# Generate definitions for new concepts using the GPT-2 model.

```python
import re
import pandas as pd
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch
from tqdm import tqdm
df=result_df
# Load your DataFrame, assuming you have a DataFrame named 'df' with 'Content' and 'head_x' columns
# Replace 'your_data.csv' with your actual data file or source
# df = pd.read_csv('your_data.csv')

# Initialize GPT-2 tokenizer and model
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")

# Define a function to generate a definition using GPT-2
def generate_definition(content, concept_name):
    # Step 1: Remove spaces between words and ".", ","
    content_cleaned = re.sub(r'\s+([.,])', r'\1', content)

    # Step 2: Split the content into a corpus by detecting "."
    corpus = content_cleaned.split(".")

    # Initialize a flag to check if definition is found
    definition_found = False

    # Step 3 and 4: Search for the concept and generate a definition using GPT-2
    for sentence in corpus:
        if concept_name in sentence:
            context = sentence.strip()

            # Generate a definition using GPT-2
            input_text = f"Define '{concept_name}': {context}"
            input_ids = tokenizer.encode(input_text, return_tensors="pt")
            with torch.no_grad():
                output = model.generate(input_ids, max_length=100, num_return_sequences=1, no_repeat_ngram_size=2)

            generated_definition = tokenizer.decode(output[0], skip_special_tokens=True)
            definition_found = True
            break

    # If no definition is found, add a message
    if not definition_found:
        generated_definition = f"No definition found for '{concept_name}'"

    return generated_definition
# Apply the function to the DataFrame and add definitions to a new column with tqdm progress bar
tqdm.pandas()
df['concept_definition'] = df.progress_apply(lambda row: generate_definition(concatenated_string,row['head_list'][0]), axis=1)
```
